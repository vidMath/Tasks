```python
df_speed.boxplot(column='speed_diff',by='functionalClass')
'''seeing at correlation matrix functional class has most impact on speed_diff. Plotting speed_diff against functionClass,
the outliers are mostly with the functionalClass = 5 (for -ve difference) and functional class = 1or for +ve difference).
So may be terrain is having higher impact on speeds here''' 
```

```python
df_speed['date'], df_speed['time'] = zip(*df_speed['timestamp'].map(lambda x: x.split(' ')))
df_speed['hour'], df_speed['min'] = zip(*df_speed['time'].map(lambda x: x.split(':')))
df_speed['hour'] = df_speed['hour'].map(lambda x: int(x))
df_speed['hour'].hist()
#checking the speed distribution based on time of the day
```
